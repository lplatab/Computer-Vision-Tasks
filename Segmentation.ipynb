```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
def load_image(path_image):
    img = cv2.imread(path_image)
    return img[...,::-1] #BGR to RGB

img = load_image('./cube2.jpg')

blur4 = cv2.bilateralFilter(img,9,75,75)

plt.figure(figsize=(10,10))

plt.subplot(325),plt.imshow(blur4),plt.title('blur4')
plt.xticks([]), plt.yticks([])
plt.show()    
```

```python
hsv = cv2.cvtColor(blur4,cv2.COLOR_RGB2HSV)
```

```python
# Segmentación de umbral de color RED

low_red = np.array([0,100,100])
high_red = np.array([2,255,255])
mask = cv2.inRange(hsv,low_red,high_red)

plt.figure(figsize=(10,10))
plt.subplot(325),plt.imshow(mask),plt.title('mask')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# Segmentación de umbral de color YELLOW
plt.figure(figsize=(10,10))
low_yellow = np.array([30,100,100])
high_yellow = np.array([45,255,255])
mask1 = cv2.inRange(hsv,low_yellow,high_yellow)
plt.subplot(325),plt.imshow(mask1),plt.title('mask1')
```

```python
# Segmentación de umbral de color ORANGE
plt.figure(figsize=(10,10))
low_orange = np.array([0,190,200])
high_orange = np.array([28,255,255])
mask2 = cv2.inRange(hsv,low_orange,high_orange)
plt.subplot(325),plt.imshow(mask2),plt.title('mask2')    ###with cube9,10 changes the range###
```

```python
# Segmentación de umbral de color WHITE
plt.figure(figsize=(10,10))
low_white = np.array([0,0,200])
high_white = np.array([145,60,255])
mask3 = cv2.inRange(hsv,low_white,high_white)
plt.subplot(325),plt.imshow(mask3),plt.title('mask3')    ###with cube7,8,12,13 changes the range###
```

```python
# Segmentación de umbral de color GREEN
plt.figure(figsize=(10,10))
low_green = np.array([40,100,100])
high_green = np.array([80,255,255])
mask4 = cv2.inRange(hsv,low_green,high_green)
plt.subplot(325),plt.imshow(mask4),plt.title('mask4')
```
