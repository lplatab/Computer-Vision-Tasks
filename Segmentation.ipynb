```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
def load_image(path_image):
    img = cv2.imread(path_image)
    return img[...,::-1] #BGR to RGB

img = load_image('./cube1.jpg')

blur4 = cv2.bilateralFilter(img,9,75,75)

plt.figure(figsize=(10,10))

plt.subplot(325),plt.imshow(blur4),plt.title('blur4')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
hsv = cv2.cvtColor(blur4,cv2.COLOR_RGB2HSV)
```

```python
# Segmentación de umbral de color red

low_red = np.array([0,100,100])
high_red = np.array([2,255,255])
mask = cv2.inRange(hsv,low_red,high_red)

plt.figure(figsize=(10,10))
plt.subplot(325),plt.imshow(mask),plt.title('mask')
plt.xticks([]), plt.yticks([])
plt.show()

```

```python
plt.figure(figsize=(10,10))
low_yellow = np.array([30,100,100])
high_yellow = np.array([45,255,255])
mask1 = cv2.inRange(hsv,low_yellow,high_yellow)
plt.subplot(325),plt.imshow(mask1),plt.title('mask1')
```

```python
plt.figure(figsize=(10,10))
low_orange = np.array([8,100,100])
high_orange = np.array([23,255,255])
mask2 = cv2.inRange(hsv,low_orange,high_orange)
plt.subplot(325),plt.imshow(mask2),plt.title('mask2')
```

```python
plt.figure(figsize=(10,10))
low_white = np.array([0,0,200])
high_white = np.array([15,13,255])
mask3 = cv2.inRange(hsv,low_white,high_white)
plt.subplot(325),plt.imshow(mask3),plt.title('mask3')
```

```python
fig, axs = plt.subplots(1,5,figsize=(20,5))

axs[0].imshow(img)
axs[0].set_title('Imagen')

fig_img = axs[1].imshow(mask)
axs[1].set_title('RED')

fig_img = axs[2].imshow(mask1)
axs[2].set_title('YELLOW')

fig_img = axs[3].imshow(mask2)
axs[3].set_title('ORANGE')

fig_img = axs[4].imshow(mask3)
axs[4].set_title('WHITE')
```
